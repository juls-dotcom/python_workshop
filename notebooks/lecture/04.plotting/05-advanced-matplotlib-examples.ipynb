__Shortcut__: instead of calling `plt.figure` and getting the axes, we can also use `plt.subplots`.

`plt.subplots` returns __a figure object__ with room for one or multiple axes and an __array with the axes objects__.

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.linspace(0, 2 * np.pi, 128, endpoint=True)
y_cos, y_sin = np.cos(x), np.sin(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y_sin);
```

We can also tell `plt.subplots` to initialize multiple axes on a figure:

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
```

And that can be really handy if we want to use loops to generate plots:

```python
from scipy.stats import norm
from random import uniform

nrows, ncols = 2, 2
fig, axes = plt.subplots(nrows=nrows, ncols=ncols)

for i in range(nrows): #or axes.flatten()
    for j in range(ncols):
        
        mean, std = uniform(-1, 1), uniform(1, 2)
        x_norm = norm.rvs(loc=mean, scale=std, size=100)
        
        axes[i, j].hist(x_norm, alpha=0.6, bins=20)
        
        t = r'$\mu = {0:.1f}, \quad \sigma = {1:.1f}$'.format(mean, std)
        axes[i, j].set_title(t)
        axes[i, j].set_xticks([-4, 0, 4])
        axes[i, j].set_yticks([])
fig.tight_layout() 
```

## The OOP API - Summary

* The OOP API allows us to manipulate plots in a more programmatic way.
* Everything is an object .
* Adjust object properties in `matplotlib` with `obj.set_*`:
    * `ax.set_xlabel`
    * `ax.set_xtitle`)
* Beginners generally start off with `plt`, but it's good to know how to do the more advanced stuff!

```python
ax = plt.axes([0, 0, 1, 1], polar=True, facecolor='white')

N = 20
theta = np.arange(0.0, 2*np.pi, 2*np.pi/N)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)
bars = plt.bar(theta, radii, width=width, bottom=0.0)

for r, bar in zip(radii, bars):
    bar.set_facecolor(plt.cm.jet(r/10.))
    bar.set_alpha(0.5)
    
ax.set_xticklabels([])
ax.set_yticklabels([]);
```
