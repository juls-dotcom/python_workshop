```python
# Say we want to drop a column from the dataframe
my_col_to_drop = None # Set column to drop
df.drop(columns={my_col_to_drop}) #Drop column
```

```python
# Now let's call df again
df
```

<div class="alert alert-block alert-info"> 
We see that <b>df</b> still have the NaN column in there! <br><br>
This is because you dropped the column but did not store the output of that operation somewhere.<br><br>
While it might seem counterintuitive at first, this makes sure that your operations are not affecting the data directly. In other words, <b>you can test your code to get the output you want, and then only save it. </b>

There are two ways one could save the results of an operation, such as dropping a column:

```python
# Either overwrite the object with the code output
df_modif = df.drop(columns={my_col_to_drop})
```

OR

```python
# use Inplace=True to modify in place
df.drop(columns={my_col_to_drop}, inplace=True)
```

<div class="alert alert-block alert-danger"> 
<b>CAUTION</b>: <br>
* Using the inplace parameter means that you will be <b>modifying the source object</b>. 
* Here, df has been modified and the column has been dropped directly in there. <br><br>
* If you want to <b>keep the past data</b>, it is better to <b>store each step in a separate objet</b>, depending on the size of your data of course ;)

* First feature: Opening Rate

Opening rate is already provided as a column. But let's verify the accuracy of that data point by calculating the opening rate given the raw data:

```python
# Assign a new column with calculated open rate
df = df.assign(open_rate_calc = df['Unique Emails Opened'] / df['Emails Sent'])
```

We just add a column with the calculated `open rate`.

Now, we can quickly plot it to look at the distribution

```python
# Sanity check to verify the distribution of the calculated open rate
(df
.open_rate_calc
.plot(kind='hist', bins=100)
)
```

<div class="alert alert-block alert-info"> 
* Note how each function is in a new row, while the overall statement is in (). <br>
* When possible, use this notation to make it easier to read your code

Since we wanted to observe the accuracy of the provided open rate, let's plot the difference between the `provided open rate` and the `calculated open rate`:

```python
# Visualize the degree of difference between the provided & calculated open rate
(df['Open Rate'] - df['open_rate_calc']).plot(kind='hist',bins=100)
```

Now you reported to management that you can some differences in `open_rate`...very small, but still. And they ask you to remove these data points because they are a bug in the system that should not eneter your analysis. 

Let's remove them using `indexing`:

<div class="alert alert-block alert-info"> 
* Indexing is based on using a boolean array (True or False) to filter out the values <br>
* it is a fast and easy to implement method to select data based on criteria & threshold

```python
# Let's create an index of the open rates that differ by > 0.01
mask = (df['Open Rate'] - df['open_rate_calc']) >= 0.01
```

```python
# How many True values fo I have?
mask.sum()
```

```python
# Let's not overwrite the data, and instead, let's store this first clean version in a separate object
df_clean = df.loc[~mask]
```
